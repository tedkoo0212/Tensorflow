```python
# input layer ~ hidden layer ~ output layer
# Boston Estate Market Example
# (503,13) ~ (503,5) ~ (503,1)
# (13 -> 5) , (5 -> 1)
#
# modeling with one hidden layer
'''

X = tf.keras.layers.Input(shape=[13])
H = tf.keras.layers.Dense(5, activation='swish')(X)
H = tf.keras.layers.Dense(3, activation='swish')(H)
H = tf.keras.layers.Dense(3, activation='swish')(H)
Y = tf.keras.layers.Dense(1)(H)

model = tf.keras.models.Model(X,Y)
model.compile(loss='mse')
'''
```

```python
# library 사용
import tensorflow as tf
import pandas as pd
```

```python
#1 과거의 데이터를 준비합니다.
파일경로 = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/boston.csv'
보스턴 = pd.read_csv(파일경로)
보스턴.columns

#1-1 독립,종속 변수 설정
독립 = 보스턴[['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax',
       'ptratio', 'b', 'lstat']]
종속 = 보스턴[['medv']]
print(독립.shape, 종속.shape)
```

```python
#2 모델의 구조를 만듭니다.
# 히든 레이어 3겹

X = tf.keras.layers.Input(shape=[13])
H = tf.keras.layers.Dense(5, activation='swish')(X)
H = tf.keras.layers.Dense(3, activation='swish')(H)
H = tf.keras.layers.Dense(3, activation='swish')(H)
Y = tf.keras.layers.Dense(1)(H)

model = tf.keras.models.Model(X,Y)
model.compile(loss='mse') #regression -> loss='mse'
```

```python
#3 모델을 학습시킵니다.
model.fit(독립,종속,epochs=1000)
```

```python
model.summary()
#한번씩 체크 해주는 것 좋다.
```

```python
# parameter : 가중치의 개수
# Output Shape : 뭐지? 왜 None?
# 
```

```python
# 4 모델을 이용합니다 (추측을 해봅니다)
print(model.predict(독립[:5]))
```

## **아이리스 품종 분류**

```python
# 1.과거의 데이터를 준비합니다.
# 2.원핫인코딩 (**)(하는 이유: Supervised Learning - Classification 이여서?)
# 3.독립 변수, 종속 변수
# 4.모델의 구조를 만듭니다.
# 5.데이터로 모델을 학습합니다.(FIT)
# 6.모델을 이용합니다.(추정해봅니다.)
```

```python
import tensorflow as tf
import pandas as pd
파일경로 = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/iris.csv'
아이리스 = pd.read_csv(파일경로)
```

```python
아이리스 = pd.get_dummies(아이리스)
print(아이리스)
print(아이리스.columns)
```

```python
독립 = 아이리스[['꽃잎길이', '꽃잎폭', '꽃받침길이', '꽃받침폭']]
종속 = 아이리스[['품종_setosa', '품종_versicolor','품종_virginica']]
print(독립.shape, 종속.shape)
```

```python
X = tf.keras.layers.Input(shape=[4])
H = tf.keras.layers.Dense(8, activation='swish')(X)
H = tf.keras.layers.Dense(8, activation='swish')(H)
H = tf.keras.layers.Dense(8, activation='swish')(H)
Y = tf.keras.layers.Dense(3, activation='softmax')(H)
model = tf.keras.models.Model(X,Y)
model.compile(loss='categorical_crossentropy', metrics='accuracy')
```

```python
model.fit(독립,종속,epochs=1000)
```

```python
print(model.predict(독립[:5]))
print(종속[0:5])
```
